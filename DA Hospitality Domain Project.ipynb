 1. What is an average occupancy rate in each of the room categories?

```python
new_col = df_aggr_bookings.apply(lambda row: row['successful_bookings']/row['capacity'], axis=1)
df_aggr_bookings = df_aggr_bookings.assign(occ_pct=new_col.values)
```
